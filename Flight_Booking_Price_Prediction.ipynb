**4. One Hot Encoding**

Performing One Hot Encoding for categorical features of a DataFrame.

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['airline'] = le.fit_transform(df['airline'])
df['source_city'] = le.fit_transform(df['source_city'])
df['departure_time'] = le.fit_transform(df['departure_time'])
df['stops'] = le.fit_transform(df['stops'])
df['arrival_time'] = le.fit_transform(df['arrival_time'])
df['destination_city'] = le.fit_transform(df['destination_city'])
df['class'] = le.fit_transform(df['class'])
df.info()
```

**5. Feature Selection**

```python
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

SELECTING THE FEATURES USING VIF.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
col_list = []
for col in df.columns:
  if((df[col].dtype != 'object') & (col != 'price')):
    col_list.append(col)

X = df[col_list]
vif_data = pd.DataFrame()
vif_data['feature'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X.values, i)
                    for i in range(len(X.columns))]
print(vif_data)
```

VIF should be less than 5 and as per above feature selection all features are having VIF less than 5.

**6. Implementing ML Algorithms**

* **Linear Regression**

```python
x = df.drop(columns=['flight','price'])
y = df['price']
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)
y_pred = pd.DataFrame(y_pred,columns=['Pred'])
df_pred = pd.DataFrame(columns=['Actual','Predicted'])
df_pred['Actual'] = y_test.reset_index()['price']
df_pred['Predicted'] = y_pred.reset_index()['Pred']
df_pred
```

Calculating r2 score, MAE, MAPE, MSE, RMSE.

```python
from sklearn.metrics import r2_score
print('r2_Score:',r2_score(y_test,y_pred))
from sklearn import metrics
print('Mean Abs Error:', metrics.mean_absolute_error(y_test,y_pred))
from sklearn.metrics import mean_absolute_percentage_error
print('Mean Abs Percentage Error:', metrics.mean_absolute_percentage_error(y_test,y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test,y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

Root Mean Squared Error (RMSE) of Linear Regression model is 7014.37 and Mean Absolute Percentage Error (MAPE) is 43 Precent. Lower the RMSE nd MAPE better the model.

PLOTING THE GRAPH OF ACTUAL AND PREDICTED PRICE OF FLIGHT.

```python
sns.distplot(y_test,label='Actual')
sns.distplot(y_pred,label='Predicted')
plt.legend()
plt.show()
```

* **Decision Tree Regression**

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
dt.fit(x_train,y_train)
y_pred = dt.predict(x_test)
print('r2_Score:',r2_score(y_test,y_pred))
print('Mean Abs Error:', metrics.mean_absolute_error(y_test,y_pred))
print('Mean Abs Percentage Error:', metrics.mean_absolute_percentage_error(y_test,y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test,y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

Mean Absolute Percentage Error is 8.3 percent and RMSE is 3785.56 which less than Linear Regression Model

PLOTING THE GRAPH OF ACTUAL AND PREDICTED PRICE OF FLIGHT.:

```python
sns.distplot(y_test,label='Actual')
sns.distplot(y_pred,label='Predicted')
plt.legend()
```

* **Random Forset Regressor**

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
rfr.fit(x_train,y_train)
y_pred =rfr.predict(x_test)
print('r2_Score:',r2_score(y_test,y_pred))
print('Mean Abs Error:', metrics.mean_absolute_error(y_test,y_pred))
print('Mean Abs Percentage Error:', metrics.mean_absolute_percentage_error(y_test,y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test,y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

Mean Absolute Percentage Error is 7.7 percent and RMSE is 2988 which is less than Linear Regression and Decision Tree Model.
